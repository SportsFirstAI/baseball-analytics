```python
from baseballcv.functions import LoadTools
from ultralytics import YOLO

print("Downloading and loading models... this takes a moment.")

load_tools = LoadTools()

# Download their specific models using the CORRECT aliases we just found
phc_path = load_tools.load_model("phc_detector")
ball_path = load_tools.load_model("ball_trackingv4") # Using their v4 ball tracker!
bat_path = load_tools.load_model("bat_tracking") # The new bat tracker!

# Load them into Ultralytics
phc_model = YOLO(phc_path)
ball_model = YOLO(ball_path)
bat_model = YOLO(bat_path)
pose_model = YOLO('yolo26l-pose.pt')

print("вњ… All 4 Models Loaded Successfully! You are clear to run Cell 3.")
```

```python
import cv2
import numpy as np

input_video = 'testVidBaseball.mp4'
output_video = 'output_baseballcv_perfect.mp4'

cap = cv2.VideoCapture(input_video)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

out = cv2.VideoWriter(output_video, cv2.VideoWriter_fourcc(*'mp4v'), fps, (width, height))

# 17-point skeleton connections
SKELETON_CONNECTIONS = [(5, 6), (5, 7), (7, 9), (6, 8), (8, 10), (5, 11), (6, 12), (11, 12), (11, 13), (13, 15), (12, 14), (14, 16)]

print("Processing frames using BaseballCV models... please wait.")
frame_count = 0
total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

contact_display_frames = 0 # Timer to keep the "CONTACT!" graphic on screen

while cap.isOpened():
    ret, frame = cap.read()
    if not ret: break
    frame_count += 1

    if frame_count % 30 == 0:
        print(f"Progress: {frame_count}/{total_frames} frames ({(frame_count/total_frames)*100:.1f}%)")

    # List to store the bounding boxes of players we want skeletons for (Pitcher and Hitter)
    target_player_boxes = []

    # 1. Run Pitcher/Hitter/Catcher Model (Filters out the crowd automatically!)
    phc_results = phc_model(frame, verbose=False, conf=0.50)

    if phc_results[0].boxes is not None:
        for box in phc_results[0].boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0].cpu().numpy())
            conf = box.conf[0].cpu().item()
            cls_id = int(box.cls[0].cpu().item())
            cls_name = phc_model.names[cls_id].lower()

            # Save pitcher and hitter coordinates for the skeleton filter
            if cls_name in ["pitcher", "hitter"]:
                target_player_boxes.append((x1, y1, x2, y2))

            # Draw exact replica of BaseballCV's bounding box + label
            cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 0, 0), 2)
            label = f"{cls_name} {conf:.2f}"
            (w, h), _ = cv2.getTextSize(label, cv2.FONT_HERSHEY_SIMPLEX, 0.6, 1)
            cv2.rectangle(frame, (x1, y1 - 25), (x1 + w, y1), (255, 0, 0), -1)
            cv2.putText(frame, label, (x1, y1 - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)

    # 2. Run Pose Model (ONLY drawn on the Pitcher and Hitter)
    pose_results = pose_model(frame, verbose=False, conf=0.50)
    if pose_results[0].keypoints is not None and len(target_player_boxes) > 0:

        for keypoints in pose_results[0].keypoints.xy.cpu().numpy():
            # Find the center of gravity of this skeleton to check if it's inside a target box
            valid_pts = [pt for pt in keypoints if pt[0] > 0]
            if not valid_pts:
                continue

            center_x = sum(pt[0] for pt in valid_pts) / len(valid_pts)
            center_y = sum(pt[1] for pt in valid_pts) / len(valid_pts)

            # Check if this skeleton belongs to one of our target players
            is_target_player = False
            for (px1, py1, px2, py2) in target_player_boxes:
                if px1 < center_x < px2 and py1 < center_y < py2:
                    is_target_player = True
                    break

            if is_target_player:
                for kp1_idx, kp2_idx in SKELETON_CONNECTIONS:
                    pt1, pt2 = keypoints[kp1_idx], keypoints[kp2_idx]
                    if pt1[0] > 0 and pt2[0] > 0:
                        cv2.line(frame, (int(pt1[0]), int(pt1[1])), (int(pt2[0]), int(pt2[1])), (0, 255, 0), 2)
                for pt in keypoints:
                    if pt[0] > 0:
                        cv2.circle(frame, (int(pt[0]), int(pt[1])), 4, (0, 0, 255), -1)

    # 3. Run Baseball Tracker v4
    ball_box_coords = None
    ball_results = ball_model(frame, verbose=False, conf=0.20)
    if ball_results[0].boxes is not None and len(ball_results[0].boxes) > 0:
        box = ball_results[0].boxes[0] # Highest confidence ball
        bx1, by1, bx2, by2 = map(int, box.xyxy[0].cpu().numpy())
        conf = box.conf[0].cpu().item()
        ball_box_coords = (bx1, by1, bx2, by2)

        # Draw clean, small box around the ball
        cv2.rectangle(frame, (bx1, by1), (bx2, by2), (0, 255, 255), 2)
        label = f"ball {conf:.2f}"
        (w, h), _ = cv2.getTextSize(label, cv2.FONT_HERSHEY_SIMPLEX, 0.5, 1)
        cv2.rectangle(frame, (bx1, by1 - 20), (bx1 + w, by1), (0, 255, 255), -1)
        cv2.putText(frame, label, (bx1, by1 - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 1)

    # 4. Run Bat Tracker
    bat_box_coords = None
    bat_results = bat_model(frame, verbose=False, conf=0.25)
    if bat_results[0].boxes is not None and len(bat_results[0].boxes) > 0:
        box = bat_results[0].boxes[0] # Highest confidence bat
        batx1, baty1, batx2, baty2 = map(int, box.xyxy[0].cpu().numpy())
        conf = box.conf[0].cpu().item()
        bat_box_coords = (batx1, baty1, batx2, baty2)

        # Draw a purple box around the bat
        cv2.rectangle(frame, (batx1, baty1), (batx2, baty2), (255, 0, 255), 2)
        label = f"bat {conf:.2f}"
        (w, h), _ = cv2.getTextSize(label, cv2.FONT_HERSHEY_SIMPLEX, 0.5, 1)
        cv2.rectangle(frame, (batx1, baty1 - 20), (batx1 + w, baty1), (255, 0, 255), -1)
        cv2.putText(frame, label, (batx1, baty1 - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 1)

    # 5. Collision Detection (Did the bat hit the ball?)
    if ball_box_coords is not None and bat_box_coords is not None:
        bx1, by1, bx2, by2 = ball_box_coords
        batx1, baty1, batx2, baty2 = bat_box_coords

        # Mathematical check for intersecting rectangles
        if bx1 < batx2 and bx2 > batx1 and by1 < baty2 and by2 > baty1:
            contact_display_frames = 15 # Set the timer for 15 frames (~0.5 sec)

    # Display "CONTACT!" Graphic
    if contact_display_frames > 0:
        text = "CONTACT!"
        (tw, th), _ = cv2.getTextSize(text, cv2.FONT_HERSHEY_DUPLEX, 2.0, 4)
        tx = (width - tw) // 2
        ty = int(height * 0.2) # Top center of the screen
        cv2.putText(frame, text, (tx, ty), cv2.FONT_HERSHEY_DUPLEX, 2.0, (0, 0, 255), 4)
        contact_display_frames -= 1

    out.write(frame)

cap.release()
out.release()
print("Processing complete! Ready for download.")
```
